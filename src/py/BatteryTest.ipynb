```python
from baselines import deepq
import baselines.common.tf_util as U
from SimpleBatteryModelDefnoCappenalty import SimpleBatterySimEnv
def test():
    tf.reset_default_graph()
    
    Lmpfile = '/home/haow889/BatteryDRL/TestData/2017_Zonal_LMP_LONGIL.csv'
    istartday =199
    
    env = SimpleBatterySimEnv(Lmpfile, istartday)
    
    #nn_model = "/home/haow889/BatteryDRL/src/py/previous_model/simplebattery_model_round1_lr_0.0001_72k.pkl"
    #nn_model = "/home/haow889/BatteryDRL/src/py/previous_model/simplebattery_model_round2_nolastpenalty_lr_0.0001_100k.pkl"
    nn_model = "/home/haow889/BatteryDRL/src/py/previous_model/simplebattery_model_nolastpenalty_multidayslast3days_lr_0.0001_100k.pkl"
    
    act = deepq.load(nn_model)
    
    with tf.variable_scope("deepq", reuse=True):
        eps = tf.get_variable("eps")
    U.set_value(eps, 0)
    print("Check epsilon is: %f" % U.eval(eps))

    #obs, done = env._validate(1,8,1.0,0.58), False
    episode_rew = 0
    #actions = np.random.randint(low = 0, high = 21, size = (3*24*2+2,1))
    observations = list()
    rewards = list()
    actions = list()
    
    istep = 0
    done = False
    obs = env._validate(istartday)
    
    while not done:
        #action = actions[istep]
        action = act(obs[None])[0]
        
        #print ('action is %d, %s'%(action, type(action)))
        
        obs, rew, done, _ = env.step(action)
        #print ('obs shape:')
        #print (obs.shape)
        observations.append(obs)   
        rewards.append(rew)
        
        episode_rew += rew
        istep += 1
    print("Episode reward for round 1:", episode_rew)       
    
    return np.array(actions), np.array(observations), np.array(rewards)
```

```python
test_actions, test_observations, test_reward = test()
```
